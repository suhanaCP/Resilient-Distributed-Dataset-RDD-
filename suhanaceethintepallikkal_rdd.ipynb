```
#cleaning the data 
clinicaltrial_2021RDD = sc.textFile("/FileStore/tables/clinicaltrial_2021.csv")
header = clinicaltrial_2021RDD.first()
clinicaltrial_2021RDD = clinicaltrial_2021RDD.filter(lambda line: line != header)
```

```
#task1-question1-RDD
distinctRDD = clinicaltrial_2021RDD.map(lambda line: line.split("|")[0]).distinct().count()
print("Distinct:", distinctRDD)
```

```
#task1-question2-RDD
type_countRDD = clinicaltrial_2021RDD.map(lambda row: (row.split('|')[5], 1)).reduceByKey(lambda x, y: x+y).sortBy(lambda x: x[1], ascending=False)

type_countRDD.collect()
```

```
#task1-question3-RDD

conditionRDD = (clinicaltrial_2021RDD
                .map(lambda row: row.split('|'))
                .filter(lambda line: line[7])
                .flatMap(lambda line: line[7].split(','))
                .map(lambda Conditions: (Conditions, 1))
                .reduceByKey(lambda x, y: x + y)
                .sortBy(lambda x: x[1], ascending=False))

conditionRDD.take(5)
```

```
#task1-quesyion4-rdd
pharmaRDD = spark.sparkContext.textFile("/FileStore/tables/pharma.csv")
header = pharmaRDD.first()
pharmaRDD = pharmaRDD.filter(lambda row: row != header)

clinicaltrial_2021_RDD = clinicaltrial_2021RDD.map(lambda line: line.split("|")).\
map(lambda x: (x[1], 1))

pharma_RDD = pharmaRDD.map(lambda line: line.split(",")) \
                      .map(lambda x: (x[1].replace('"', ''))) \
                      .map(lambda o: (o,1))

joined_rdd = clinicaltrial_2021_RDD.leftOuterJoin(pharma_RDD).filter(lambda x: x[1][1] == None)

sponsor_count_rdd = joined_rdd.map(lambda x: (x[0], 1)).reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1],ascending=False)

sponsor_count_rdd.take(10)
```

```
#task1-question5-rdd
Complete_StudiesRDD = clinicaltrial_2021RDD.filter(lambda x: '2021' in x.split('|')[4] and 'Completed' in x)

StudiesRDD = Complete_StudiesRDD.map(lambda x: (x.split('|')[4][:3], 1)).reduceByKey(lambda x, y: x + y).sortBy(lambda x: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'].index(x[0]))

StudiesRDD.collect()
```
